```python
import matplotlib.pyplot as plt
import pandas as pd

#  Defining  the performance data
data = {
    "Model": ["Dummy", "KNN-1", "KNN-2", "KNN-3-1", "KNN-3-2"],
    "Accuracy": [0.5075, 0.895, 0.963, 0.721, 0.965],
    "Recall": [0.5048, 0.8867, 0.9407, 0.4163, 0.913],
    "Precision": [0.5327, 0.9045, 0.9061, 1.0, 0.8873]
}

# Converting  the data dictionary into a pandas DataFrame
df = pd.DataFrame(data)

# figure size
plt.figure(figsize=(12, 6))

# metrics
metrics = ["Accuracy", "Recall", "Precision"]

# metrics
for metric in metrics:
    plt.plot(df["Model"], df[metric], marker='o', label=metric)

# plotting the graph
plt.title(" Performance Comparison (Excl. Train Accuracy)")
plt.xlabel("Model") #labels
plt.ylabel("Score") # labels
plt.ylim(0, 1.1)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)

#displaying the plot
plt.show()
```

Dummy:	Baseline performance; all metrics are low,
KNN-1:	Strong overall performance; high accuracy, recall, and precision. This model has good balance
KNN-2:	Even better performance than KNN-1, especially in recall. This model is excellent for general use Excellent for general
KNN-3-1:	Strange behavior perfect precision of 1.0 but very low recall 0.4. Likely predicts only one class correctly and misses most others.
KNN-3-2: Similar to KNN-2, with very high accuracy and recall, and slightly better balance than KNN-3-1.
